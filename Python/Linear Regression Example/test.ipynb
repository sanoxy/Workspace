```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

advertising = pd.read_csv("Company_data.csv")
advertising
```

```python
advertising.shape
advertising.info()
advertising.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(advertising,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',size = 4,aspect = 1, kind='scatter')
plt.show()
```

```python
advertising.corr()
```

```python
sns.heatmap(advertising.corr(), cmap='PuBuGn', annot = True)
```

```python
X = advertising['TV']
y = advertising['Sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
X_train
y_test
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
```

```python
lr = sm.OLS(y_train, X_train_sm).fit()

lr.params
```

```python
lr.summary()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 6.948 + 0.054 * X_train, 'r')
plt.show()
```

```python
y_train__pred = lr.predict(X_train_sm)

res = (y_train - y_train__pred)
```

```python
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms',fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
X_test_sm = sm.add_constant(X_test)
```

```python
y_test_pred = lr.predict(X_test_sm)
y_test_pred
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y_test,y_test_pred)
r_squared
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_test_pred, 'r')
plt.show()
```

```python
r2_score(y_train,y_train__pred)
r2_score(y_test,y_test_pred)
```
